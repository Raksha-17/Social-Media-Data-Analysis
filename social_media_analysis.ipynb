## Task 1 – Import required libraries

As the name suggests, the first step is to import all the necessary libraries that will be used in the project. In this case, we need pandas, numpy, matplotlib, seaborn, and random libraries.

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Task 2 – Generate random data for the social media data

```python
# 1) Define the categories
categories = ["Food", "Travel", "Fashion", "Fitness", "Music", "Culture", "Family", "Health"]

# 2) Choose the length of the data
n = 500  # you can adjust this as needed

# 3) Generate the Date column using a date range
dates = pd.date_range(start="2021-01-01", periods=n)

# 4) Generate the Category column with random choices
category_column = [random.choice(categories) for _ in range(n)]

# 5) Generate the Likes column with random integers in [0, 10000)
likes_column = np.random.randint(0, 10000, size=n)

# 6) Assemble the data dictionary
data = {
    "Date": dates,
    "Category": category_column,
    "Likes": likes_column
}

# 7) Load into a DataFrame
df = pd.DataFrame(data)

#8) Display the first few rows
print("Head of DataFrame:")
print(df.head(), "\n")
```

## Task 3 – Load the data into a Pandas DataFrame and Explore the data

```python
# 1) Load into a DataFrame
df = pd.DataFrame(data)

# 2) Display the first few rows
print("Head of DataFrame:")
print(df.head(), "\n")

# 3) DataFrame info (dtypes, non-null counts, memory usage)
print("DataFrame Info:")
print(df.info(), "\n")

# 4) Descriptive statistics for numeric columns
print("DataFrame Description (numeric):")
print(df.describe(), "\n")

# 5) Count of each Category
print("Category Counts:")
print(df['Category'].value_counts(), "\n")
```

## Task 4 – Clean the data

```python
# 1) Remove null data (drop rows with any nulls)
df_clean = df.dropna()

# 2) Remove duplicate rows
df_clean = df_clean.drop_duplicates()

# 3) Convert the 'Date' field to datetime using pandas on the Series
# If 'Date' is already datetime-like, this is still safe.
df_clean['Date'] = pd.to_datetime(df_clean['Date'])

# 4) Ensure 'Likes' is an integer type
df_clean['Likes'] = df_clean['Likes'].astype(int)

# Optional: reset index after cleaning
df_clean = df_clean.reset_index(drop=True)

# Display some confirmation
print("Cleaned DataFrame info:")
print(df_clean.info())
print("\nFirst few rows after cleaning:")
print(df_clean.head())
```

## Task 5– Visualize and Analyze the data

```python
# 1) Histogram of Likes
plt.figure(figsize=(8, 5))
sns.histplot(data=df_clean, x='Likes', bins=30, kde=True)
plt.title('Histogram of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 2) Boxplot: Category vs Likes
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_clean, x='Category', y='Likes')
plt.title('Boxplot of Likes by Category')
plt.xlabel('Category')
plt.ylabel('Likes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3) Mean of the 'Likes' column (overall)
mean_likes = df_clean['Likes'].mean()
print(f"Overall mean of Likes: {mean_likes:.2f}")

# 4) Mean of Likes per Category using groupby
mean_by_category = df_clean.groupby('Category')['Likes'].mean().sort_values(ascending=False)
print("\nMean Likes by Category:")
print(mean_by_category)

# Optional: display a quick summary
print("\nSummary:")
print(df_clean.describe(include='all'))
```
